```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
name=['sepal-length','sepal-width','petal-length','petal-width','Class']
i1= pd.read_csv(r'C:\Users\Utkarsh\Desktop\Iris_training_dataset.csv',names= name)
i1
```

```python
x= i1[['sepal-length','sepal-width','petal-length','petal-width']]
y= i1[['Class']]
```

```python
from sklearn.model_selection import *
from sklearn.metrics import *
from sklearn.neighbors import KNeighborsClassifier
```

```python
teach= KNeighborsClassifier()
k= list(range(1,31))
para_grid= dict(n_neighbors= k)

kfold= StratifiedKFold(10)
grid= GridSearchCV(teach,para_grid, cv=kfold.split(x,y), scoring='accuracy')
grid.fit(x,y)
```

```python
grid_mean_scores= [result.mean_validation_score for result in grid.grid_scores_]
plt.plot(k,  grid_mean_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross Validation Accuracy')
plt.grid()
```

```python
dataset= pd.read_csv(r'C:\Users\Utkarsh\Desktop\iris_prediction_dataset.csv', names= name)
x_test= dataset[['sepal-length','sepal-width','petal-length','petal-width']]
```
